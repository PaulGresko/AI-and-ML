**Вариант: alcohol, ash, alcalinity_of_ash, hue**

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import classification_report
```

```python
wine = load_wine(as_frame=True)
X = wine.data[['alcohol', 'ash', 'alcalinity_of_ash', 'hue']]
y = wine.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=52)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train_scaled, y_train)
```

```python
rfc = RandomForestClassifier(max_depth=1, random_state=52)
rfc.fit(X_train_scaled, y_train)
```

```python
param_grid = {
    'n_estimators': [10, 25, 50, 100],
    'max_depth': [None, 1, 2, 3, 4, 5, 6, 7, 8]
}

rfc_gs = GridSearchCV(RandomForestClassifier(random_state=52), param_grid, cv=5)
rfc_gs.fit(X_train_scaled, y_train)
```

```python
for i, model in enumerate([dtc, rfc, rfc_gs]):
    print("+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")
    print(f'{model.__class__.__name__}:')
    predictions = model.predict(X_test_scaled)
    report = classification_report(y_test, predictions)
    print(report)
```

 - Точность дерева решений: 0,83
 - Точность случайного леса: 0,81
 - Точность оптимального случайного леса: 0,92

Точности случайного леса и дерева решений приблезительно одинаковые относительно оптимального случайного леса.  
**Вывод: важно подобрать опатимальные гиперпараметры, что сильно улучшает точность.**
